```
# First, let's import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```
# Load the dataset from the provided link

url = 'https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv'

data = pd.read_csv(url)

```

```
# Display first few rows of the dataset to get an idea of the data

print(data.head())

```

```
# Check for null values in the dataset

print(data.isnull().sum())

```

```
# Generating descriptive statistics of the dataset

print(data.describe())

```

```
# Plotting the distribution of each column using histogram

data.hist(bins=10, figsize=(20,15))
plt.show()

```

```
# Plotting the distribution of each column using boxplot

data.boxplot(figsize=(20,15))
plt.show()

```

```
# Plotting pairwise relationships between different columns of the dataset

sns.pairplot(data, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=7, aspect=0.7, kind='reg')
plt.show()

```

```
# Analyzing correlation coefficients between different columns of the dataset

corr = data.corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.show()

```
